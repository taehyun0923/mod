```python
data_1.head()
```

```python
# '마약범죄에 해당하는 행만 선택
data_2 = data_1[data_1['범죄대분류'] == '마약범죄']

data_2
```

```python
data_3 = data_2.iloc[0, 2:] # 0번 열이 범죄대분류, 1번 열이 범죄중분류, 2번 열부터 끝 열까지 서울시 자치구별 마약범죄 발생 건수이므로 2번 열부터 데이터를 추출
data_3
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as format
import matplotlib as mpl
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic' 
```

```python
plt.figure(figsize=(16, 6))
data_3.plot(kind='bar') # 세로 막대기 그래프를 사용
plt.title('서울시 자치구별 마약범죄 발생 건수', fontsize=16)
plt.xlabel('자치구')
plt.ylabel('발생 건수')
plt.xticks(rotation=45)
plt.show()
```
